```python
import acopy
import networkx as nx

# --- Ciudades y distancias ---
cities = [0, 1, 2, 3, 4]
distances = [
    [0, 10, 15, 20, 25],
    [10, 0, 35, 25, 30],
    [1, 35, 0, 30, 20],
    [2, 2, 30, 0, 15],
    [5, 3, 20, 15, 0]
]

# creacion grafo
G = nx.Graph()
G.add_nodes_from(cities)

# ahora se agregan las aristas solo entre ciudades distintas
for i in cities:
    for j in cities:
        if i < j:  # evita duplicados en grafo no dirigido
            weight = distances[i][j]
            G.add_edge(i, j, weight=weight)

# --- Config ACOpy ---
solver = acopy.Solver()
colony = acopy.Colony(alpha=1, beta=3)

tour = solver.solve(G, colony, limit=100)

print("Mejor ruta:", tour.nodes)
print("Costo total:", tour.cost)
```
